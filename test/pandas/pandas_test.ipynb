# Pandas可视化

## plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum() #新的ts数组值为之前ts之前数组值之和

ts.plot()#以ts为y轴，index为x轴
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

#matplotlib.style.use('ggplot')

df = pd.DataFrame(np.random.randn(1000, 4), index=pd.date_range('20170101', periods=1000), columns=list('ABCD'))
df = df.cumsum()
df.plot()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

df3 = pd.DataFrame(np.random.randn(1000, 10), columns=['B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K']).cumsum()
df3['A'] = pd.Series(list(range(len(df))))
df3.plot(x='A')
plt.show()
```

## Bar

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_bar = pd.DataFrame(np.random.randn(10, 4), columns=['A', 'B', 'C', 'D']).cumsum()
df_bar.iloc[5].plot(kind='bar')
df_bar.plot.bar(stacked=True)
df_bar.plot.bar()
df_bar.plot.barh()
df_bar.plot.barh(stacked=True)
plt.axhline(0, color='k')
plt.show()
```

## Histograms


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_hist = pd.DataFrame({'a':np.random.randn(10)+1, 'b':np.random.randn(10),
                       'c':np.random.randn(10)-1}, columns=['a', 'b', 'c'])
df_hist.plot.hist(alpha=0.5, stacked=True)
df_hist.plot.hist(alpha=0.5, stacked=True, orientation='horizontal', bins=20)
plt.show()
```

## box

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_box = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])
df_box.plot.box()
plt.show()
```

## area

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_area = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])
df_area.plot.area()
df_area.plot.area(stacked=False)

plt.show()

```

## scatter

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_scatter = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])
ax = df_scatter.plot.scatter(x='a', y='b', color='DarkGreen', label='Group 1')
df_scatter.plot.scatter(x='c', y='d', color='DarkRed', label='Group 2', ax=ax, s=df_scatter['c']*200)
plt.show()
```

## hexagonal bin

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_hex = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df_hex['b'] += np.arange(1000)
df_hex.plot.hexbin(x='a', y='b', gridsize=25)
plt.show()
```

## pie

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

series_pie = pd.Series(3*np.random.rand(4), index=['a', 'b', 'c', 'd'], name='series')
series_pie.plot.pie(figsize=(6, 6))

df_pie = pd.DataFrame(3*np.random.rand(4, 4), index=['a', 'b', 'c', 'd'], columns=['x', 'y', 'z', 't'])
df_pie.plot.pie(labels=['AA', 'BB', 'CC', 'DD'], colors=['r', 'g', 'b', 'c'], subplots=True, figsize=(16, 16), layout=(2, 2), autopct='%.2f')
plt.show()
```
